```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import collections as c
```

```python
url1 = "okcupid_profiles.csv"
okcupid_df = pd.read_csv(url1) 
okcupid_df
```

```python
okcupid_df = okcupid_df.dropna()
okcupid_df.isnull().sum()
okcupid_df.reset_index().copy()
```

```python
okcupid_df.columns = okcupid_df.columns.str.replace(' ', '')
okcupid_df['age'].astype(int)
```

```python
okcupid_df['sex'].hist()
```

```python
c.Counter(okcupid_df['status']).most_common(20)
df = pd.DataFrame.from_records(c.Counter(okcupid_df['status']).most_common(10), columns=['status','count_status'])

df[['sex','count_sex']] = pd.DataFrame.from_records(c.Counter(okcupid_df['sex']).most_common(20), columns=['page','count'])
df[['pets','count_pets']] = pd.DataFrame.from_records(c.Counter(okcupid_df['pets']).most_common(20), columns=['page','count'])
df[['body_type','count_body_type']] = pd.DataFrame.from_records(c.Counter(okcupid_df['body_type']).most_common(20), columns=['page','count'])
df[['drugs','count_drugs']] = pd.DataFrame.from_records(c.Counter(okcupid_df['drugs']).most_common(20), columns=['page','count'])
df[['drinks','count_drinks']] = pd.DataFrame.from_records(c.Counter(okcupid_df['drinks']).most_common(20), columns=['page','count'])
# df[['height','count_height']] = pd.DataFrame.from_records(c.Counter(okcupid_df['height']).most_common(50), columns=['page','count'])
df
```

```python
df_to_plot = df.drop(columns=['status','sex', 'pets','body_type','drugs','drinks'])
df_to_plot.plot(subplots=True, layout=(5,4), kind='box', figsize=(12,14), patch_artist=True)
plt.subplots_adjust(wspace=0.5);
```
